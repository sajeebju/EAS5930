```python
import pandas as pd
import numpy as np

#make dataset
earthquake= pd.DataFrame({
    "PHI_E" : [0,0,0,0,40,40,40,60,60,60],
    "LAMBDA_E" : [0,0,0,-30,0,0,0,0,0,0],
    "PHI_S": [1,5,10,60,41,45,50,61,65,70],
    "LAMBDA_S": [1,5,10,60,1,5,10,1,5,10]

})

delX = np.array(111.195*(earthquake["PHI_S"] - earthquake["PHI_E"]))
delY = np.array(111.195*(earthquake["LAMBDA_S"] - earthquake["LAMBDA_E"])* 
                np.cos(np.radians((earthquake["PHI_E"] + earthquake["PHI_S"])/ 2)))


# distance, azimuth, back azimuth to be included as columns of the earthquake dataset
dist_km = np.sqrt(np.power(delX, 2) + np.power(delY, 2))
azimuth = np.degrees(np.arctan(delY/delX))
backAzimuth = np.degrees(np.arctan(delY/delX)) + 180

#assign new columns
earthquake['Dist(km)'] = dist_km
earthquake['Azim'] = azimuth
earthquake['backAzimuth'] = backAzimuth
earthquake
```

```python
#Input epicenter and station coordinates and get the results

phi_e = int(input("Enter the epicenter latitude : "))
lambda_e = int(input("Enter the epicenter longitude : "))
phi_s = int(input("Enter the station latitude : "))
lambda_s = int(input("Enter the station longitude : "))

delX = 111.195 * (phi_s - phi_e)
delY = 111.195 * (lambda_s - lambda_e) * np.cos(np.radians((phi_e + phi_s) / 2))


dist_km = np.sqrt(np.power(delX, 2) + np.power(delY, 2))
Az = np.degrees(np.arctan(delY/delX))
Baz = np.degrees(np.arctan(delY/delX)) + 180

# Testing for the first row
print("Epicentral distance {:.2f} km".format(dist_km))
print("Azimuth = {:.2f} degree".format(Az))
print("Back azimuth = {:.2f} degree".format(Baz))
```
