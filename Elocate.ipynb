```python
import pandas as pd
import numpy as np

max_data= pd.read_csv("/Users/ahasan/Courses/Final1a/20210913/ML/depmax.csv")
min_data= pd.read_csv("/Users/ahasan/Courses/Final1a/20210913/ML/depmin.csv")
max_data.drop("DEPMIN", axis = 1, inplace=True)
min_data.drop("DEPMAX", axis = 1, inplace = True)
max_data
```

```python
dep_max = max_data["DEPMAX"].to_numpy()
dep_min = min_data["DEPMIN"].to_numpy()

import numpy as np

# Example arrays
array1 = dep_max
array2 = dep_min

# Vectorized operation to get the maximum values
max_values_arr = np.maximum(array1, array2)

print(np.array(max_values_arr))
```

```python
ML = np.log10(max_values_arr) + 1.11 * np.log10(max_data["DISTANCE"].to_numpy()) + 0.00189* max_data["DISTANCE"].to_numpy() - 2.09
ML
```

```python
np.mean(ML)
```

```python
import numpy as np
import matplotlib.pyplot as plt
magnitude_values = ML
plt.hist(magnitude_values, bins='auto', color='blue', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')


# Show the plot
plt.show()
```

```python
np.random.seed(0) 
magnitude_values = ML

bin_counts, bins = np.histogram(magnitude_values, bins=10)
bin_centers = 0.5 * (bins[1:] + bins[:-1])

# Create bar plot
plt.bar(bin_centers, bin_counts, width=np.diff(bins), edgecolor='black', align='center', alpha=0.7)
plt.title('Bar Plot')
plt.xlabel('Magnitude')
plt.ylabel('Count')
plt.show()
```
