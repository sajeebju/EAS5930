```python
PATH_P_RAY = "/Users/ahasan/Documents/seismology_software/TAUP/p_time_ray.csv"
df_P_RAY = pd.read_csv(PATH_P_RAY)
df_P_RAY.dropna(inplace = True)
df_P_RAY
```

```python
import csv
import math

csv_file = "/Users/ahasan/Documents/seismology_software/TAUP/p_time_ray.csv"
target_ray_parameter = 6.9

closest_distance = None
closest_travel_time = None
closest_ray_parameter = float('inf')  

with open(csv_file, mode='r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  

    for row in csv_reader:
        if len(row) >= 3 and all(row[:3]):
            distance, ray_parameter, travel_time = map(float, row[:3])
            if abs(ray_parameter - target_ray_parameter) < abs(closest_ray_parameter - target_ray_parameter):
                closest_distance = distance
                closest_travel_time = travel_time
                closest_ray_parameter = ray_parameter

if closest_distance is not None and closest_travel_time is not None:
    print(f"Closest Distance: {closest_distance} degrees")
    print(f"Closest Travel Time: {closest_travel_time} seconds")
    print(f"Closest Ray Parameter: {closest_ray_parameter}")
else:
    print(f"No data found for a ray parameter close to {target_ray_parameter}")

```
